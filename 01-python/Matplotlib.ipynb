**Mathplotlib**

* It is a low level graph plotting library in python that is used for `Data Visualization`.
* It is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

**1. Basics**

* It can be installed using `pip install matplotlib`
* Version can be determined using `matplotlib.__version__`
* Most of the plotting utilities lies under the `pyplot` submodule i.e usually reffered as `plt`. This is done using `import matplotlib.pyplot as plt` or `from matplotlib import pyplot as plt`.
* If we do not specify the points in the `x-axis`, they will get the default values `0, 1, 2, 3, ..`
* Markers can only be ploted by indicating 3rd parameter or `marker='o'` - means 'rings'. More details are [here](https://www.w3schools.com/python/matplotlib_markers.asp)
* The `subplot()` is used to draw multiple plots. It accepts 3 arguments [row, column, plot sequence]

```python
import matplotlib
print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,4,5]

plt.xkcd()
plt.plot(x,y)
plt.plot(x)
plt.plot(y, "*:r")
plt.grid()
plt.show();
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,4,5]

plt.style.use("ggplot")
plt.figure(figsize=(17.4,5))
plt.title("Sample Graph",color="red")
plt.xlabel("X-Axis",color="red")
plt.ylabel("Y-Axis",color="red")

plt.subplot(1, 3, 1)
plt.plot(x,'D',linewidth=5, linestyle="solid", color="black")

plt.subplot(1, 3, 2)
plt.plot(y,'*',linewidth=5, linestyle="dashdot", color="green")

plt.subplot(1, 3, 3)
plt.plot(x,y, color="red", linewidth=5, linestyle="dashed", label="Sample Legend")


plt.legend()
plt.show();
```

**2. Bar Charts**

* `Pyplot.bar()` can be used to draw vertical bar graphs.
* `Pyplot.barh()` can be used to draw horizontal bar graphs.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,3,4,5]

plt.figure(figsize=(17.4,5))
plt.subplot(1, 3, 1)
plt.bar(x,y, width=0.5)

plt.subplot(1, 3, 3)
plt.barh(x,y,height=0.5)

plt.show();
```
